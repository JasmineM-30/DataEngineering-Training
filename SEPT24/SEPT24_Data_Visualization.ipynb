# **Sales Data Visualization**

**Prepare the Dataset**

```python
import pandas as pd

# Create sales dataset
data = {
"Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct",
"Nov", "Dec"],
"Revenue": [12000, 15000, 17000, 14000, 18000, 22000, 24000, 20000, 23000, 25000,
27000, 30000],
"Profit": [3000, 3500, 4000, 3200, 4500, 6000, 6500, 5800, 6200, 7000, 7500,
8000],
"Region": ["North", "South", "East", "West", "North", "South", "East", "West",
"North", "South", "East", "West"]
}
df = pd.DataFrame(data)
print(df.head())
```

**Exercises (Visualization Tasks)**

**1. Line Chart**
Plot Revenue over Months.
Add labels and title.

```python
import matplotlib.pyplot as plt
df.plot(x="Month", y="Revenue", kind="line", marker='o')
plt.title("Monthly Revenue")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.grid(True)
plt.show()
```

**2. Bar Chart**
Plot Profit for each Month.
Use different colors.

```python
colors = plt.cm.Paired(range(len(df)))
df.plot(x="Month", y="Profit", kind="bar")
plt.title("Monthly Profit")
plt.xlabel("Month")
plt.ylabel("Profit")
plt.show()
```

**3. Grouped Bar Chart**
Compare Revenue vs Profit per Month.

```python
df.plot(x="Month", y=["Revenue", "Profit"], kind="bar")
plt.title("Revenue vs Profit by Month")
plt.xlabel("Month")
plt.ylabel("Amount")
plt.show()
```

**4. Pie Chart**
Show total Revenue contribution by Region.

```python
region_sales = df.groupby("Region")["Revenue"].sum()
region_sales.plot(kind="pie", autopct='%1.1f%%', startangle=90, legend=True)
plt.title("Total Revenue Contribution by Region")
plt.ylabel("")  # Hides the y-label
plt.show()
```

**5. Scatter Plot**
Plot Revenue vs Profit to check correlation.

```python
df.plot(kind="scatter", x="Revenue", y="Profit")
plt.title("Revenue vs Profit")
plt.xlabel("Revenue")
plt.ylabel("Profit")
plt.show()
```

**6. Histogram**
Show the distribution of Revenue values.

```python
df["Revenue"].plot(kind="hist", bins=5, edgecolor='black')
plt.title("Distribution of Revenue")
plt.xlabel("Revenue")
plt.show()
```

**7. Box Plot**
Visualize Profit across Regions.

```python
df.boxplot(column="Profit", by="Region")
plt.title("Profit by Region")
plt.suptitle("")
plt.xlabel("Region")
plt.ylabel("Profit")
plt.show()
```
